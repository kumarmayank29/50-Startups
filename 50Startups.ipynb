**This dataset has data collected from New York, California and Florida about 50 business Startups "17 in each state". The variables used in the dataset are Profit, R&D spending, Administration Spending, and Marketing Spending.**

```
import pandas as pd
import numpy as np
```

```
dataset=pd.read_csv("50Startup.csv")
```

```
dataset.head()
```

**Check For NULL Values**


```
dataset.isnull().any()
```

```
dataset.head()
```

**Encode Values which are given as text Using Label Encoder**

```
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
dataset['State']=le.fit_transform(dataset['State'])
```

```
dataset.head()
```

```
dataset['State'].value_counts()
```

**Apply One Hot Encoding**

Even before that divide this dataset into input and output 

```

x  = dataset.iloc[:,0:4].values
y = dataset.iloc[:,4:5].values
```

```

from sklearn.preprocessing import OneHotEncoder
one = OneHotEncoder()
z=one.fit_transform(x[:,3:4]).toarray()
x=np.delete(x,3,axis=1)
x=np.concatenate((z,x),axis=1)
```

```
x #Converted to 6 columns after one hot encodimng of columns State
```

```
x.shape
```

**Split the Data**

```
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

**ANN Starts**

```
from keras.models import Sequential
from keras.layers import Dense
```

```
x_train.shape
```

```
#Declare the model
model=Sequential()
#Add input Layer
model.add(Dense(units=6,activation='relu'))
#Add hidden Layers
model.add(Dense(units=12,activation='relu'))
#Add the Output Layer
model.add(Dense(units=1,activation='relu'))
```

```
model.compile(optimizer='adam',loss='mse',metrics=["mse"])
```

**Fit the Model**

```
model.fit(x_train,y_train,epochs=700,batch_size=8)
```

```
y_pred=model.predict(x_test)
```

**Check on Random Value**

```
y_p=model.predict(np.array([[1,0,0,12345,13456,12345]]))
```
